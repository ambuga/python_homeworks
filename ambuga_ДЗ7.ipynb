```python
def first(x):
    return x**3

x = linspace(-10, 10)  
y = first(x)

def second(t):
    return 2*t**3

t = linspace(-10, 10)  
z = second(t)

print(x,y)
print(t,z)
```

```python
plt.plot(x, y, label=r"$y = x^3$")
plt.plot(t, z, label=r"$y = 2*x^3$")
plt.fill_between(x,y,z,t, color='yellow', alpha=0.5,label='place between')
plt.legend(loc = 2)
```

```python
def first(x):
    return x**2

x = linspace(-10, 10)  
y = first(x)

def second(t):
    return 2*t**2

t = linspace(-10, 10)  
z = second(t)

print(x,y)
print(t,z)

def third(a):
    return a**3
a = linspace(-10, 10)
b = first(a)

def fourth(c):
    return 2*c**3
c = linspace (-10, 10)
d = fourth(c)

print(a,b)
print(c,d)
```

```python
plt.subplot(211)
plt.plot(x, y, label=r"$y = x^2$")
plt.plot(t, z, label=r"$y = 2*x^2$")
plt.fill_between(x,y,z,t, color='yellow', alpha=0.5)
plt.legend(loc = 2)
plt.title('First')

plt.subplot(212)
plt.plot(a, b, label=r"$y = x^3$")
plt.plot(c, d, label=r"$y = 2*x^3$")
plt.fill_between(a,b,d,c, color='yellow', alpha=0.5)
plt.legend(loc = 2)
plt.title('Second')
```
